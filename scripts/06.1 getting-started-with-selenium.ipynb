# üåô About Selenium

## What's Selenium?
Selenium is a software testing framework used for automating web browsers (such as Chrome, Firefox, Safari, etc.).

Selenium is widely used for application testing and data scraping.

## How does it work?

Selenium works by providing a set of libraries and tools that interact with web browsers just like a human user would. Here's an overview of the process:

1. **WebDriver:** Selenium uses a component called WebDriver, which acts as a bridge between your code and the web browser you want to automate. It is essentially a translator between you and the browser. More accurately, WebDriver is an interface in the Selenium library. Every major web browser has its own specific WebDriver. For example:
- ChromeDriver for Google Chrome
- GeckoDriver for Mozilla Firefox
- EdgeDriver for Microsoft Edge
- ... and so on.

2. **Coding:** You write code using a programming language (in this case, Python), to tell Selenium what actions to perform. These actions can include opening a web page, clicking buttons, filling out forms, and scraping data.

3. **Execution:** When you run your code, it sends instructions to the WebDriver, which then communicates with the web browser (like Chrome, Firefox, etc.) to carry out the actions you coded.

# üë©üèª‚Äçüíª Installing Selenium

To be able to use Selenium, you first have to

1. **Install selenium in your system.** This can be most easily achieved by opening your command line / terminal and executing the `pip install selenium` command

2. **Install webdriver-manager in your system.** This can be most easily achieved by opening your terminal and executing the `pip install webdriver-manager` command.

**If you have any problem during the installation, please email the TA**

Once you have successfully completed the two steps above, it is easy to use selenium, as follows:


```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException

# create a browser object-- this should open a chrome browser using selenium
browser = webdriver.Chrome()
```

If you would like to use a different browser (not chrome), find the correct command below and use it instead:


```python
# browser = webdriver.Firefox()
# browser = webdriver.Safari()
# browser = webdriver.Edge()
```

If you've done everything correctly, a new Chrome window should open! This one you do not control by clicking on, but rather *through the Python code you're writing here!*

- Sometimes you might run into problems where your operating system will not allow you to operate your browser through automated software -- Selenium.
- E.g., when you try to open Safari with Selenium, you may get `SessionNotCreatedException: Message: Could not create a session: You must enable the 'Allow Remote Automation' option in Safari's Develop menu to control Safari via WebDriver.`
